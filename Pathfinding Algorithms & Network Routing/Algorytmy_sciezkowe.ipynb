`````python
# zad 1

from typing import Iterable
import networkx as nx
from networkx import Graph

def liscie(G: Graph) -> Iterable:
    # raise NotImplementedError("brak kodu")
    for g in G.nodes():
      if G.degree[g] == 1:
        yield g
`````

### Zad 2

Algorytm Dijkstry jest jednym z najbardziej znanych algorytmów służących do znajdowania najkrótszej ścieżki w grafie z wagami nieujemnymi. Wagi krawędzi reprezentują pewien koszt przejścia między wierzchołkami (np. odległość, czas).

Idea algorytmu Dijkstry polega na iteracyjnym rozszerzaniu obszaru poszukiwań, zaczynając od wierzchołka początkowego. W każdym kroku wybieramy wierzchołek, który ma najmniejszą przypisaną odległość od wierzchołka początkowego i rozważamy wszystkich jego sąsiadów. Jeśli odległość do sąsiada jest mniejsza niż obecnie znana, aktualizujemy ją. Proces ten powtarzamy, aż wszystkie wierzchołki zostaną odwiedzone.

Pseudokod algorytmu:

Inicjalizacja: Ustaw `d(s) = 0`, a dla każdego innego wierzchołka v ustaw `d(v) = ∞`. Ustaw `S = ∅`.

Dopóki S nie zawiera wszystkich wierzchołków:

Wybierz wierzchołek `u`, który nie należy do `S`, a dla którego `d(u)` jest minimalne.
Dodaj `u` do `S`.
Dla każdej krawędzi `(u, v)` wychodzącej z wierzchołka `u`:
Jeśli `d(u) + w(u, v) < d(v)`, to uaktualnij `d(v) = d(u) + w(u, v)` i `p(v) = u`.
Po zakończeniu algorytmu `d(v)` zawiera długość najkrótszej ścieżki od wierzchołka `s` do wierzchołka `v`, a `p(v)` zawiera poprzednika wierzchołka `v` na tej ścieżce.

Działająca implementacja w języku Python, wykorzystująca strukturę grafu z biblioteki NetworkX:

```python
# własna implementacja algorytmu Dijkstry
# wagi krawędzi zapamiętane w `weight`
import networkx as nx
from typing import Tuple

def dijkstra(G: nx.Graph, source) -> Tuple[dict, dict]:
    d = {v: float('inf') for v in G.nodes} # <1>
    d[source] = 0
    p = {v: None for v in G.nodes} # <2>
    S = set()

    while len(S) < len(G.nodes):
        u = min((v for v in G.nodes if v not in S), key=lambda v: d[v]) # <3>
        S.add(u)
        for v in G.neighbors(u):
            if d[u] + G[u][v]['weight'] < d[v]:
                d[v] = d[u] + G[u][v]['weight']
                p[v] = u

    return d, p

```

`````python
# zad 2

import networkx as nx
from typing import Tuple

def dijkstra_with_path(G, source, target)->Tuple[float, list]|None:
    d = {v: float('inf') for v in G.nodes} # <1>
    d[source] = 0
    p = {v: None for v in G.nodes} # <2>
    S = set()

    while len(S) < len(G.nodes):
        u = min((v for v in G.nodes if v not in S), key=lambda v: d[v]) # <3>
        S.add(u)
        for v in G.neighbors(u):
            if d[u] + G[u][v]['weight'] < d[v]:
                d[v] = d[u] + G[u][v]['weight']
                p[v] = u
    # Warunki z zadania 2
    if d[target] == float('inf'):
      return None

    # print(d)
    distance = d[target]

    curr_node = target

    path = []
    while curr_node != None:
      path.append(curr_node)
      curr_node = p[curr_node]

    path.reverse()

    return distance, path
`````

## Zad 3

Załóżmy, że mamy sieć drogową przedstawioną jako graf ważony. Wagi krawędzi reprezentują odległości między miastami. Chcemy znaleźć najkrótszą ścieżkę między dwoma wybranymi miastami (długość ścieżki obliczana jest jako suma wag odwiedzanych ścieżek), ale z dodatkowym ograniczeniem: możemy przejechać przez maksymalnie `k` miast (włącznie z początkowym i końcowym).

Zaproponuj algorytm, który rozwiąże ten problem.

Specyfikacja funkcji
`def dijkstra_with_limit(G: nx.Graph, source, target, limit) -> Tuple[float, list]:`

gdzie:

  - `G` - graf ważony, bez krawędzi wielokrotnych i pętelek
  - `source` - wierzchołek początkowy
  - `target` - wierzchołek końcowy
  - `limit` - maksymalna liczba miast, przez które możemy przejechać, włącznie z początkowym i końcowym
  - zwraca krotkę `(distance, path)`, gdzie:
    - `distance` - długość najkrótszej ścieżki (długość ścieżki obliczana jest jako suma wag odwiedzanych ścieżek)
    - `path` - lista wierzchołków tworzących najkrótszą ścieżkę

Dodatkowe założenia:

  - Wagi krawędzi są nieujemne.
  - W przypadku, gdy nie istnieje ścieżka spełniająca warunek, zwróć None.
  - W przypadku, gdy istnieje więcej niż jedna ścieżka spełniająca warunek, zwróć
    - jeśli istnieje więcej niż jedna najkrótsza ścieżka, zwróć tę o najmniejszej liczbie miast
    - jeśli istnieje więcej niż jedna ścieżka o najmniejszej liczbie miast, zwróć dowolną z nich.

`````python
import networkx as nx
from typing import Tuple
import heapq

def dijkstra_with_limit(G: nx.Graph, source, target, limit) -> Tuple[float, list]:
    pq = []
    heapq.heappush(pq, (0, 1, source, [source]))
    while pq:
      curr_dist, curr_cities, u, curr_path =  heapq.heappop(pq)
      if u == target:
        return (curr_dist, curr_path)
      elif curr_cities == limit:
        continue

      for v in G.neighbors(u):
        heapq.heappush(pq,
          ((curr_dist + G[u][v]['weight']),
            curr_cities+1,
            v,
            curr_path + [v]
          ))


    return None
`````

## Zad 4

Algorytm Bellmana-Forda służy do znajdowania najkrótszych ścieżek w grafie skierowanym z wagami krawędzi. Algorytm ten jest wolniejszy od algorytmu Dijkstry, ale radzi sobie z ujemnymi wagami krawędzi.

Idea działania algorytmu Bellmana-Forda

1. Inicjalizacja:
  - Każdemu wierzchołkowi przypisujemy odległość równą nieskończoności, z wyjątkiem wierzchołka początkowego, któremu przypisujemy wartość 0.
  - Tworzymy tablicę poprzedników, która będzie śledzić ścieżkę do każdego wierzchołka.
2. Relaksacja krawędzi:
  - Pętlę główną wykonujemy |V| - 1 razy, gdzie |V| jest liczbą wierzchołków w grafie.
  - W każdej iteracji przechodzimy przez wszystkie krawędzie grafu i sprawdzamy, czy możemy skrócić odległość do wierzchołka docelowego danej krawędzi. Jeśli tak, aktualizujemy odległość i poprzednika.
3. Wykrywanie ujemnych cykli:
  - Po zakończeniu głównej pętli przechodzimy jeszcze raz przez wszystkie krawędzie. Jeśli znajdziemy jakąkolwiek krawędź, której relaksacja skutkuje zmniejszeniem odległości, oznacza to, że w grafie istnieje ujemny cykl osiągalny z wierzchołka początkowego.

Implementacja (uproszczona) w języku Python
```python
from collections import defaultdict

def bellman_ford(graph, source):                
    distance = defaultdict(lambda: float('inf'))
    distance[source] = 0                         

    # Relaksacja krawędzi
    for _ in range(len(graph) - 1):             
        for u in graph:                          
            for v, weight in graph[u]:
                if distance[u] != float('inf') and distance[u] + weight < distance[v]:
                    distance[v] = distance[u] + weight

    # Sprawdzenie ujemnych cykli
    for u in graph:                             
        for v, weight in graph[u]:
            if distance[u] != float('inf') and distance[u] + weight < distance[v]:
                return None  # Istnieje ujemny cykl

    return distance  
```

`````python
import networkx as nx

def bellman_ford(G, source):
    # distance = defaultdict(lambda: float('inf'))
    distance  = {v: float('inf') for v in G.nodes}
    distance[source] = 0
    p = {v: None for v in G.nodes}
    S = set()

    for _ in range(len(G.nodes) - 1):
      for u in G.nodes:
        for v in G.neighbors(u):
          if distance[u] != float ('inf') and distance[u] + G[u][v]['weight'] < distance[v]:
            distance[v] = distance[u] +  G[u][v]['weight']

    for u in G.nodes():
      for v in G.neighbors(u):
        if distance[u] != float('inf') and distance[u] + G[u][v]['weight'] < distance[v]:
          return None

    return distance
`````
